```python
#Cuantos filas son peliculas y cuantas son shows de televisión
movies = data[data.type == "Movie"]
tv_shows = data[data.type == "TV Show"]
print(len(movies))
print(len(tv_shows))
```

```python
#Agrupacion del dataframe por rating
grouped = data.groupby("rating")

for name, group in grouped:
    print(name, len(group))
```

```python
#Agrupacion del dataframe por rating y por tipo(Pelicula o show de television)
grouped2 = data.groupby(["rating", "type"])

for name, group in grouped2:
    print(name, len(group))
```

```python
# Recuperamos la información de un único grupo de interés.
grouped2.get_group(("PG-13", "Movie"))
```
