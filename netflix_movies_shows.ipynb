```python
import datetime
import re

#Dado que la columna date_added tiene fechas no informadas(NaN) se porcede a cambiar este
#valor por una fecha de defecto 31-Dec-99

data.fillna({'date_added': "31-Dec-99"}, inplace=True)

#Se compureba que los valores NaN de dicha columna han sido cambiados por la fecha
#por defecto 31-Dec-99

data[data.date_added == "31-Dec-99"]

#Se cambia el formato de fecha por el formato yyyy-mm-dd. Se analiza tres tipos de formatos distintos:
# dd-mes-yyy(01-Sep-17), Mes d, yyyy(November 1, 2019) y Mes dd, yyyy(November 10, 2019)
for i in data["date_added"]:
     if re.findall(r'\w{2}-\w+-\w{2}', i):
            datetime_from_str = str(datetime.datetime.strptime(i, "%d-%b-%y"))
            datetime_from_str2 = datetime_from_str.replace(" 00:00:00","")
            print(datetime_from_str2)
            
     elif re.findall(r' \d{1},', i):
        d = str(re.findall(r'\d{1},', i))
        d2 = re.sub("\d{1},", "0"+d[2]+",", i)
        datetime_from_str3 = str(datetime.datetime.strptime(d2, " %B %d, %Y"))
        datetime_from_str4 = datetime_from_str3.replace("00:00:00","")
        print(datetime_from_str4)
        
     else :
        datetime_from_str5 = str(datetime.datetime.strptime(i, " %B %d, %Y"))
        datetime_from_str6 = datetime_from_str5.replace("00:00:00","")
        print(datetime_from_str6)
```

```python
#Agrupacion del dataframe por clasificación
grouped = data.groupby("rating")

#cuantos hay para cada clasificación
for name, group in grouped:
    print(name, len(group))
```

```python
#Agrupacion del dataframe por clasificación y por tipo(Pelicula o show de television)
grouped2 = data.groupby(["rating", "type"])

for name, group in grouped2:
    print(name, len(group))
```

```python
#La variable genres es un array(de una dimension) de listas
genres = pd.unique(data["genres"])

#Se crea una lista para guardar los generos unicos
list_genres = []

#Se itera cada elemento del array convirtiendolo en una lista de listas

for i in list(genres):
    
    # Se separa cada elemento de la lista por coma y espacios
    y = i.split(sep=", ")
    
    #Se itera sobre cada elemento separado
    for x in y:
        
        #Se verifica si el elemento esta en la lista list_genres, si no esta, se agrega
        if x not in list_genres:
            list_genres.append(x)
            
print("Hay {} generos distintos que son:".format(len(list_genres)))

for g in list_genres:
    print(g)
```
